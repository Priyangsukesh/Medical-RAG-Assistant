```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_community.llms import HuggingFacePipeline

model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Build HF pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512,do_sample=False,eos_token_id=tokenizer.eos_token_id)

# Wrap it for Langchain
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# 1. Split CSV documents (already in `docs`) into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
split_docs = text_splitter.split_documents(docs)

# 2. Define embedding model (any small one for speed/efficiency)
embedding_model = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2",
    #model_kwargs={"device": "cuda"},
    encode_kwargs={"normalize_embeddings": True}
)
# 3. Create FAISS vector store for retrieval
db = FAISS.from_documents(split_docs, embedding_model)
```

```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("medquad.csv")  # Update the path if needed

# Check column names
print(df.columns)

# Limit to 10 rows before converting  # Gets the first 10 rows
df_sample=df.head(3282)
# Create test data in required format
test_data = [
    {"question": row["question"], "answer": row["answer"]}
    for _, row in df_sample.iterrows()
]
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.tokenize import word_tokenize
from tqdm import tqdm
import numpy as np
import re

# --- Normalization ---
def normalize(text):
    text = text.strip().lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

# --- Token-level Metrics ---
def compute_token_metrics(predicted, ground_truth):
    pred_tokens = normalize(predicted).split()
    gt_tokens = normalize(ground_truth).split()

    common = set(pred_tokens) & set(gt_tokens)
    num_same = sum(min(pred_tokens.count(tok), gt_tokens.count(tok)) for tok in common)

    precision = num_same / len(pred_tokens) if pred_tokens else 0
    recall = num_same / len(gt_tokens) if gt_tokens else 0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0

    total_tokens = max(len(pred_tokens), len(gt_tokens))
    accuracy = num_same / total_tokens if total_tokens else 0

    return precision, recall, f1, accuracy

# --- Lists to collect metrics ---
bleu_scores = []
exact_matches = []
token_precisions = []
token_recalls = []
token_f1s = []
token_accuracies = []

smooth_fn = SmoothingFunction().method1

# --- Evaluation Loop ---
for item in tqdm(test_data):
    question = item["question"]
    ground_truth = item["answer"]
    
    predicted = answer_with_top_doc(question)

    # BLEU
    bleu = sentence_bleu(
        [word_tokenize(ground_truth)],
        word_tokenize(predicted),
        smoothing_function=smooth_fn
    )
    bleu_scores.append(bleu)

    # Exact Match
    exact_match = normalize(predicted) == normalize(ground_truth)
    exact_matches.append(1 if exact_match else 0)

    # Token-level Metrics
    precision, recall, f1, accuracy = compute_token_metrics(predicted, ground_truth)
    token_precisions.append(precision)
    token_recalls.append(recall)
    token_f1s.append(f1)
    token_accuracies.append(accuracy)

# --- Final Results ---
print("\n📊 Evaluation Metrics (Token-Level):")
print(f"Exact Match        : {np.mean(exact_matches):.4f}")
print(f"BLEU Score         : {np.mean(bleu_scores):.4f}")
print(f"Token-level Accuracy : {np.mean(token_accuracies):.4f}")
print(f"Token-level Precision: {np.mean(token_precisions):.4f}")
print(f"Token-level Recall   : {np.mean(token_recalls):.4f}")
print(f"Token-level F1 Score : {np.mean(token_f1s):.4f}")
```
